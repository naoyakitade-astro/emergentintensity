# Public fitting interface for emergent Stokes I

This code describes the emergent Stokes $I$ from a slab with a given inclination using Equation (17) of Kitade & Kataoka (2026).  
The fitting coefficients $(A_I, B_I, I_{\rm conv}, \omega_I)$ are modeled as polynomial functions of $\omega$, and fourth-order polynomials are adopted in this implementation.

Note that for very large inclinations, the fitting formula may fail to reproduce the numerical solution accurately. We therefore recommend using this formula only for inclinations below approximately 80 degrees.

```python
from stokes_i_fitting import fit_and_plot_I
from stokes_i_fitting import evaluate_I_from_fit_result
from stokes_i_fitting import load_I_only_inp
```

```python
# Show the fitting formula, comparison with numerical results, and relative error.
# Load the tables explicitly so repeated calls do not rely on module-level global state.
data = load_I_only_inp("data/StokesI_emergent")

inc = 45.0  # inc: inclination angle in degrees
result = fit_and_plot_I(inc=inc, data=data)
```

```python
# Evaluate the fitted Stokes I value at any (omega, tau_max)
omega = 0.5
tau_max = 1.0
i_value = evaluate_I_from_fit_result(omega=omega, tau_max=tau_max, fit_result=result)
print(i_value)
```
