# Public fitting interface for emergent polarization fraction

This code describes the emergent polarization fraction $PF=Q/I$ from a slab with a given inclination using Equation (18) of Kitade & Kataoka (2026).   
The fitting coefficients $(A_{\rm PF}, B_{\rm PF}, PF_{\rm conv}, \tau_{\rm PF})$ are modeled as polynomial functions of $\omega$, and fifth-order polynomials are adopted in this implementation.
The value of $PF$ obtained from the fitting formula corresponds to $Q/I$, while the plotted quantity is shown in percent, i.e., $100\times PF$.

Note that for very large inclinations, the fitting formula may fail to reproduce the numerical solution accurately. We therefore recommend using this formula only for inclinations below approximately 80 degrees.

```python
from pf_fitting import fit_and_plot_PF
from pf_fitting import evaluate_PF_from_fit_result
from pf_fitting import load_IQ_inp
```

```python
# Show the fitting formula, comparison with numerical results, and relative error.
# Load the tables explicitly so repeated calls do not rely on module-level global state.
data = load_IQ_inp(
    indir_I="data/StokesI_emergent",
    indir_Q="data/StokesQ_emergent",
)

inc = 45.0  # inc: inclination angle in degrees
result = fit_and_plot_PF(inc=inc, data=data)
```

```python
# Evaluate the fitted PF value at any (omega, tau_max)
omega = 0.5
tau_max = 1.0
pf_value = evaluate_PF_from_fit_result(omega=omega, tau_max=tau_max, fit_result=result)
print(pf_value)
```
