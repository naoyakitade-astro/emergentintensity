# Example usage

This notebook uses the new split-format data files:
- `StokesI_emergent/I_tau..._omega....inp`
- `StokesQ_emergent/Q_tau..._omega....inp`
- `Q_table.inp`

If these files are not located under the default `data/` directory, call `setup_tables(...)` explicitly.


```python
from stokes_interpolation import emergent_stokes, setup_tables

# Optional but recommended for repeated evaluations:
# explicitly load the interpolation context once and pass it around.
context = setup_tables(data_dir="data")

# Example
tau_max = 1.0  # Vertical extinction optical depth
omega = 0.5  # Dust albedo
inc = 45.0  # Inclination in degrees

I_ex, Q_ex = emergent_stokes(
    tau_max=tau_max,
    omega=omega,
    inc=inc,
    context=context,
)

print("\nExample (emergent_stokes):")
print(f"  I({tau_max}, {omega}, {inc}deg)  = {float(I_ex):.6e}")
print(f"  Q({tau_max}, {omega}, {inc}deg)  = {float(Q_ex):.6e}")
print(f"  Q/I = {float(Q_ex / I_ex) if I_ex != 0.0 else 0.0:.6e}")
```
